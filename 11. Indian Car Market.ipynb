# 1. Filter cars that have a price greater than 2,000,000

```python
expensive_car = df[df['Price']<2000000]

expensive_car
```

# 2. Find the average mileage for each fuel type

```python
avg_mileage = df.groupby('Fuel_Type')['Mileage'].mean()

avg_mileage
```

# 3. Plot a histogram to visualize the distribution of car prices

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Price'], bins = 10, kde = True)

plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Car Price Distribution")
plt.show()
```

# 4. Show a heatmap of correlations between numerical columns

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot = True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

# 5. Scatter plot to see the relationship between Toyota's price and mileage

```python
toyota_data = df[df['Brand']=='Toyota']

plt.figure(figsize=(8, 5))
sns.scatterplot(x='Mileage', y='Price', data = toyota_data, hue = 'Fuel_Type')
plt.xlabel("Mileage (km/l)")
plt.ylabel("Price")
plt.title("Mileage vs Price")
plt.show()
```

# 6. Bar chart showing the average service cost per car brand

```python
avg_service_cost = df.groupby('Brand')['Service_Cost'].mean()

plt.figure(figsize=(8, 5))
sns.barplot(x=avg_service_cost.index, y=avg_service_cost.values)
plt.xticks(rotation=45)
plt.xlabel("Brand")
plt.ylabel("Average Service Cost")
plt.title("Average Service Cost by Brand")
plt.show()
```

# 7. Most expensive car for each brand in the dataset

```python
most_expensive = df.loc[df.groupby('Brand')['Price'].idxmax()]
most_expensive[['Brand', 'Model', 'Price']]
```

# 8. Count Cars by Transmission Type

```python
transmission_counts = df["Transmission"].value_counts()
print(transmission_counts)

plt.figure(figsize=(6, 4))
sns.barplot(x=transmission_counts.index, y=transmission_counts.values, palette="viridis")
plt.xlabel("Transmission Type")
plt.ylabel("Count")
plt.title("Number of Cars by Transmission Type")
plt.show()
```
